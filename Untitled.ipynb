```python
df_curves = pd.read_csv("CrossValidationEvaluation_Curves.csv")
df_params = pd.read_csv("CrossValidationEvaluation_Params.csv")
input_curves = df_curves["Input curves"].to_numpy().reshape(1000, 150)
predicted_curves = df_curves["Predicted curves"].to_numpy().reshape(1000, 150)
df_curves.head()
```

```python
t = np.load("t.npy")
plt.plot(t, input_curves[78], label="Input from fold $F_{1}$")
plt.plot(t, predicted_curves[78], label="Prediction from model $M_{2345}$")
plt.title("$\chi^{2}$-test is inf")
plt.legend()
plt.show()
```

```python
import math
chisq_list = []
for i in range(1000):
    
    # Perform the chi-squared test for each curve pair
    # chisq = scipy.stats.chi2_contingency(np.abs(predicted_curves[i][1:]), np.abs(input_curves[i][1:]))
    chisq = ((predicted_curves[i][1:] - input_curves[i][1:])**2 / input_curves[i][1:])
    print(np.sum(chisq))
    if(math.isnan(np.sum(chisq)) == False and math.isinf(np.sum(chisq)) == False):
        chisq_list.append(np.sum(chisq))
    print(i)
```

```python
chisq = np.array(chisq_list)
print(chisq[chisq < 100000].shape)
plt.figure(figsize=(10, 10))
plt.title(f"$\chi^{2}$-test for network fitting of size: {len(chisq)}")
plt.xlabel("$\chi^{2}$ value")
plt.ylabel("Counts")
plt.hist(chisq, bins = 100)
plt.show()
```
